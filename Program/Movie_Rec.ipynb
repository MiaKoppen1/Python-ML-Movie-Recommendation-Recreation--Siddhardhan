```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from difflib import get_close_matches
from difflib import *

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#PRE-PROCESSING
```

```python
movie_data = pd.read_csv('C:/Users/user/Desktop/Personal Projects/Python ML Movie Recommendation/Data/movies.csv')
```

```python
#CHOSING THE FEATURES FOR THE RECOMMENDATION SERVICE.
```

```python
selected_ft = ['genres', 'keywords', 'tagline', 'cast', 'director'] 
print(selected_ft)
```

```python
#FILLING ALL NULL VALUES IN THE DATASET.
```

```python
for feature in selected_ft: 
    movie_data[feature] = movie_data[feature].fillna('')
```

```python
#COMBINING THE FEATURES.
```

```python
combined_ft = movie_data['genres'] + ' '+ movie_data['keywords'] + ' ' + movie_data['tagline'] + ' ' + movie_data['cast'] + ' ' + movie_data['director']
```

```python
print(combined_ft)
```

```python
#CONVERTING TEXT TO VECTORS.
```

```python
v1 = TfidfVectorizer()
vector_ft = v1.fit_transform(combined_ft)  #converting the data to feature vectors.
```

```python
#DATA ANALYSIS
```

```python
#RETRIEVING THE SIMILARITY SCORE.
```

```python
sim = cosine_similarity(vector_ft)

print(sim) #showing similarity with cosine_similarity.
```

```python
print(sim.shape)
```

```python
#USER INPUTS THEIR FAVORITE MOVIE.
```

```python
word = input('What is Your Favorite Movie? :')
possibilities = movie_data['title'].tolist()
print(possibilities)
```

```python
#FINDING SIMILAR MOVIE/S.
```

```python
get_similar_match = difflib.get_close_matches(word, possibilities)
print(get_similar_match)
```

```python
#FINDING CLOSE MATCHES.
```

```python
find_close_match = difflib.get_close_matches(word, possibilities)
print(find_close_match)

close_match = find_close_match[0]
print(close_match)
```

```python
#CREATING AN INDEX OF MOVIES THAT ARE CLOSE MATCHES.
```

```python
index_of_the_movie = movie_data[movie_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#RETRIEVING SIMILAR MOVIES.
```

```python
similarity_score = list(enumerate(sim[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#SORTING THE MOVIES BASED ON THEIR SIMILARITY
```

```python
sorted_similar_movies = sorted(sim, key = lambda x:x[1], reverse = True) 
```

```python
#PRINTING OUT A LIST OF RECOMMENDED MOVIES TO THE USER.
```

```python
print('Recommended Movies for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
